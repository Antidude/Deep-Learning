```python
import os
import zipfile
import csv
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.utils as ku

class mycallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self,epoch,logs = {}):
    if((logs.get('acc')>.996 and logs.get('val_acc')>.996)):
      self.model.stop_training = True

callbacks = mycallback()
```

```python

pixels = []
labels = []
with open("/kaggle/input/Kannada-MNIST/train.csv", 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)
    for row in reader:
        labels.append(row[0])
        pixel = row[1:]
        pixels.append(pixel)
```

```python
pixel2D = []
for pixel in pixels:
  pixel = np.array(pixel).reshape(28,28)
  pixel2D.append(pixel)
```

```python
x = np.expand_dims(labels,axis = -1)

print(np.array(pixels).shape)
z = np.concatenate((x,np.array(pixels)),axis = 1)
np.array(z).shape
```

```python
z = z.astype("int32")
```

```python
labels = z[:,0]
pixels = z[:,1:] 
```

```python
pixel2D = []
for pixel in pixels:
  pixel = np.array(pixel).reshape(28,28)
  pixel2D.append(pixel)
```

```python
training_images = pixel2D[:50000]
training_labels = labels[:50000]
validation_images = pixel2D[50000:]
validation_labels = labels[50000:]
```

```python
training_images = np.expand_dims(training_images,axis = -1)
validation_images = np.expand_dims(validation_images,axis = -1)
```

```python
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(#featurewise_center=True, 
                             #featurewise_std_normalization=True,
                            rescale = 1./255
                            #shear_range = .1,
                             zoom_range = .1,
                            width_shift_range = 3,
                            height_shift_range = 3,
                            rotation_range = 20,
                            #zca_whitening = True, 
                            fill_mode = "nearest"
                            )

copy_train =  training_images.copy()
Dict = {'theta': 5,'tx' : 5,'ty' : 5,'zx' : .1,'zy':.1,'brightness' : 1.2}


#copy_train[0] = datagen.apply_transform(copy_train[0],
#                        Dict)
#training_images = np.append(training_images,copy_train, axis=0)
#print(training_images.shape)
#s = np.sum(training_images[1]-p[1])
#print(np.abs(s))
datagen.fit(training_images)
```

```python
p = datagen.flow(copy_train,training_labels,shuffle=True,batch_size = 55000)
```

```python
new_images = np.array(p[0][0])
new_labels = np.array(p[0][1])

print(new_images.shape)
print(new_labels.shape)
```

```python
print(training_images[0].shape) 
```

```python
model1 = tf.keras.Sequential([ 
    tf.keras.layers.BatchNormalization(axis=-1, 
                                    momentum=0.99, 
                                    epsilon=0.001, 
                                    center=True, 
                                    scale=True, 
                                    beta_initializer='zeros', 
                                    gamma_initializer='ones', 
                                    moving_mean_initializer='zeros', 
                                    moving_variance_initializer='ones', 
                                    beta_regularizer=None, 
                                    gamma_regularizer=None, 
                                    beta_constraint=None, gamma_constraint=None,
                                    input_shape = (28,28,1)),
                                                    
    tf.keras.layers.Conv2D(256,(3,3),activation = "relu"),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "softmax")
]) 

model2 = tf.keras.Sequential([
    tf.keras.layers.BatchNormalization(axis=-1, 
                                    momentum=0.99, 
                                    epsilon=0.001, 
                                    center=True, 
                                    scale=True, 
                                    beta_initializer='zeros', 
                                    gamma_initializer='ones', 
                                    moving_mean_initializer='zeros', 
                                    moving_variance_initializer='ones', 
                                    beta_regularizer=None, 
                                    gamma_regularizer=None, 
                                    beta_constraint=None, gamma_constraint=None,
                                    input_shape = (28,28,1)),

    tf.keras.layers.Conv2D(256,(3,3),activation = "relu"),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "softmax")
]) 

model3 = tf.keras.Sequential([
    tf.keras.layers.BatchNormalization(axis=-1, 
                                    momentum=0.99, 
                                    epsilon=0.001, 
                                    center=True, 
                                    scale=True, 
                                    beta_initializer='zeros', 
                                    gamma_initializer='ones', 
                                    moving_mean_initializer='zeros', 
                                    moving_variance_initializer='ones', 
                                    beta_regularizer=None, 
                                    gamma_regularizer=None, 
                                    beta_constraint=None, 
                                    gamma_constraint=None,
                                    input_shape = (28,28,1)),

    tf.keras.layers.Conv2D(256,(3,3),activation = "relu"),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    tf.keras.layers.Conv2D(64,(3,3),activation = "relu"),
    #tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.15),
    
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(32,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "softmax")
]) 
```

```python
model1.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"]
             ) 
model2.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"]
             ) 
model3.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"]
             ) 
```
